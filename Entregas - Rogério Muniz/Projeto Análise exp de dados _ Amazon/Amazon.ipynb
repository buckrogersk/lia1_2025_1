## 5. Tratamento de atributos com valores ausentes
##### Verificação de atributos com valores ausentes na database

```python
# índices das linhas que contém valores NaN
idxRowNan = pd.isnull(df_dataset).any(axis=1).to_numpy().nonzero()

# imprime apenas as linhas com valores ausentes
display(df_dataset.iloc[idxRowNan])
```

<div>
Será preenchido os valores vazios com a média de contagem de avaliação tendo de acordo a avaliação (rating)

```python
def trataFaltantes(df_dataset):

    # Garantir que 'rating_count' seja numérico
    df_dataset['rating_count'] = pd.to_numeric(df_dataset['rating_count'], errors='coerce')

    # Calcular a média de 'rating_count' para cada categoria de 'rating'
    media_por_rating = df_dataset.groupby('rating')['rating_count'].mean()

    # Substituir os valores nulos pela média correspondente do 'rating'
    df_dataset['rating_count'] = df_dataset.apply(
        lambda row: media_por_rating[row['rating']] if pd.isnull(row['rating_count']) else row['rating_count'], axis=1
    )

    return df_dataset

# Aplicando a função
df_dataset = trataFaltantes(df_dataset)

# Verificando se ainda há valores nulos
print('\nValores nulos restantes:')
print(df_dataset.isnull().sum())
```

## 7. Análise dos Dados
A análise das características presentes em um conjunto de dados permite a descoberta de padrões e tendências que podem fornecer informações valiosas para compreender o processo que gerou os dados.
Contudo as colunas numéricas estão sendo interpretadas como strings devido à presença de símbolos. <div>
Para que o describe() funcione corretamente, precisamos converter esses valores para números.

```python
# Função para remover símbolos e converter para número
def limpar_numerico(valor):
    if isinstance(valor, str):  # Verifica se o valor é uma string
        valor = valor.replace('₹', '').replace('%', '').strip()  # Remove símbolos
        return pd.to_numeric(valor, errors='coerce')  # Converte para número
    return valor  # Retorna o valor original se não for string

# Aplicar a conversão nas colunas relevantes
colunas_para_converter = ['discounted_price', 'actual_price', 'discount_percentage', 'rating']
for coluna in colunas_para_converter:
    df_dataset[coluna] = df_dataset[coluna].apply(limpar_numerico)

# Agora, podemos visualizar as estatísticas corretamente
df_detalhes = df_dataset.describe()
display(df_detalhes)
```

## 8. Detecção de outliers (valor atípico)
Outro passo importante na análise e tratamento dos dados é a detecção de outliers (dados gerados por leituras incorretas, erros de digitação etc).

- Gráfico Box Plot

```python
# gerar um bloxplot para cada atributo
df_dataset.boxplot(figsize=(15,7))
plt.show()
```

---
## 9. Remoção de Outliers

```python
# Função para remover outliers usando o método IQR
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Lista das colunas numéricas para remoção de outliers
num_columns = ["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]

# Aplicar remoção de outliers
df_cleaned = remove_outliers(df_dataset, num_columns)
```

---
## 10. Geração do gráfico

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df_cleaned[num_columns])
plt.xticks(rotation=45)
plt.title("Distribuição de preços, descontos, avaliações e contagens de rating após limpeza")
plt.show()
```
