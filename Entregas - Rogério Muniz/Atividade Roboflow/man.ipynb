### Execução do modelo treinado em um vídeo:

```python
from ultralytics import YOLO
import cv2

# Carregar o modelo
model = YOLO("robo_env/Scripts/runs/detect/train2/weights/best.pt")  

# Caminho do vídeo de entrada
video_path = "Utils/video1.mp4"
cap = cv2.VideoCapture(video_path)

# Obter informações do vídeo
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS)

# Criar objeto para salvar vídeo de saída
out = cv2.VideoWriter("Utils/output_video.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# Processar quadro a quadro
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Rodar a inferência
    results = model(frame)[0]  # resultado do primeiro frame
    annotated_frame = results.plot()  # adicionar anotações

    # Escrever no vídeo de saída
    out.write(annotated_frame)

# Liberar recursos
cap.release()
out.release()
print("✅ Vídeo salvo como: Utils/output_video.mp4")
```

### Execução do modelo utilizando uma webcam 

```python
from ultralytics import YOLO
import cv2

# Carregar o modelo
model = YOLO("robo_env/Scripts/runs/detect/train2/weights/best.pt")

# Abrir webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Fazer predição com confiança mínima de 0.5
    results = model.predict(source=frame, conf=0.65, verbose=False)

    # Desenhar detecções no frame
    annotated_frame = results[0].plot()

    # Mostrar frame
    cv2.imshow("YOLO Webcam", annotated_frame)

    # Pressione Q para sair
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
